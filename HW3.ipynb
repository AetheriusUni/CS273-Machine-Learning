Part 1. The data for class 0 vs 1 is linearly separable since you can draw a line from the bottom-left to the top-right in between the data points without any of the green points being in the region with red points and vice versa. The data for class 1 vs 2 is not linearly separable since there is no way a line can be drawn in between the data points without any of the green points being in the region with red points and vice versa.

```python
import sys
import numpy as np
import mltools as ml
import matplotlib.pyplot as plt

iris = np.genfromtxt("data/iris.txt",delimiter=None)
X, Y = iris[:,0:2], iris[:,-1] # get first two features & target
X,Y = ml.shuffleData(X,Y) # reorder randomly (important later)
X,_ = ml.transforms.rescale(X) # works much better on rescaled data

#XA = X[x values which are matched with Y < 2 AKA class 0&1 , corresponding y values]
XA, YA = X[Y<2,:], Y[Y<2] # get class 0 vs 1
XB, YB = X[Y>0,:], Y[Y>0] # get class 1 vs 2

plt.title('Class 0 v 1')
#dataset1, dataset2, paints points according to classes in YA, colormap
plt.scatter(XA[:, 0], XA[:, 1], c = YA, cmap = 'prism')
plt.show()

plt.title('Class 1 v 2')
#dataset1, dataset2, paints points according to classes in YB, colormap
plt.scatter(XB[:, 0], XB[:, 1], c = YB, cmap = 'prism')
plt.show()
```
